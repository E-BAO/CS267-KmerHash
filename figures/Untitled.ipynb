```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for single node and multiple procs

# naive
naive_insert_time_test = [0.370594] * 7
naive_insert_time_large = [2.923044] * 7
naive_insert_time_human = [9.915974] * 7

naive_total_time_test = [2.151551] * 7
naive_total_time_large = [16.695905] * 7
naive_total_time_human = [56.778476] * 7

proc = [1, 2, 4, 8, 16, 32, 64]
# parallel
insert_time_test = [2.144266, 1.169603, 0.638707, 0.405937, 0.311282, 0.189602, 0.204113]
insert_time_large = [15.265685, 8.773247, 4.683528, 3.051949, 2.282866, 1.896889, 2.314290]
insert_time_human = [52.253949, 29.655163, 17.958581, 11.712502, 8.381051, 6.367426, 8.408266]

total_time_test = [5.767438, 3.183410, 1.720767, 1.049488, 0.882063, 0.914132, 1.350313]
total_time_large = [40.829169, 22.801048, 12.725866, 7.418117, 4.857841, 3.351654, 3.934321]
total_time_human = [138.514090, 77.490083, 43.619819, 26.679091, 17.009317, 10.651061, 13.132597]
```

```python
# compute strong scale efficiency
scale_eff_insert_test = [naive_insert_time_test[0] / (2**i * insert_time) * 100 for i, insert_time in enumerate(insert_time_test)]
scale_eff_insert_large = [naive_insert_time_large[0] / (2**i * insert_time) * 100 for i, insert_time in enumerate(insert_time_large)]
scale_eff_insert_human = [naive_insert_time_human[0] / (2**i * insert_time) * 100 for i, insert_time in enumerate(insert_time_human)]

plt.figure()
plt.plot(proc, scale_eff_insert_test, marker='*', linewidth=0.8, label='test')
plt.plot(proc, scale_eff_insert_large, marker='*', linewidth=0.8, label='large')
plt.plot(proc, scale_eff_insert_human, marker='*', linewidth=0.8, label='human')
my_x_ticks = np.arange(0, 70, 4)
my_y_ticks = np.arange(0, 22, 1)
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)

plt.xlabel("# proc")
plt.ylabel("Strong scale efficiency/%")
plt.legend(loc='best')
plt.grid(True)
plt.savefig("scale_single_insert.png")
```

```python
scale_eff_total_test = [naive_total_time_test[0] / (2**i * total_time) * 100 for i, total_time in enumerate(total_time_test)]
scale_eff_total_large = [naive_total_time_large[0] / (2**i * total_time) * 100 for i, total_time in enumerate(total_time_large)]
scale_eff_total_human = [naive_total_time_human[0] / (2**i * total_time) * 100 for i, total_time in enumerate(total_time_human)]

plt.figure()
plt.plot(proc, scale_eff_total_test, marker='*', linewidth=0.8, label='test')
plt.plot(proc, scale_eff_total_large, marker='*', linewidth=0.8, label='large')
plt.plot(proc, scale_eff_total_human, marker='*', linewidth=0.8, label='human')
my_x_ticks = np.arange(0, 70, 4)
my_y_ticks = np.arange(0, 44, 2)
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)

plt.xlabel("# proc")
plt.ylabel("Strong scale efficiency/%")
plt.legend(loc='best')
plt.grid(True)
plt.savefig("scale_single_total.png")
```
