```python
# time for serial, naive gpu and gpu
n = np.log([500, 1000, 2000, 4000, 8000])
time_serial = np.log([0.557748, 2.20084, 8.7491, 35.3984, 147.749])
time_naive_gpu = np.log([0.199916, 0.360529, 0.681752, 1.35951, 2.76885])
time_our_gpu = np.log([0.977607, 0.983489, 1.18899, 1.20491, 1.80752])
```

```python
plt.figure()
plt.plot(n, time_serial, marker='*', linewidth=0.5, label='serial')
plt.plot(n, time_naive_gpu, marker='*', linewidth=0.5, label='naive_gpu')
plt.plot(n, time_our_gpu, marker='*', linewidth=0.5, label='our_gpu')

my_x_ticks = np.arange(6, 10, 0.5)
my_y_ticks = np.arange(-1, 6, 0.5)
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)

plt.xlabel('Number of particles')
plt.ylabel('Time/s')
plt.grid(True)
plt.legend(loc='best')
plt.savefig('gpu_time.png')
plt.show()
```
